```python
import pandas as pd
```

### Unemployment 

```python
une = pd.read_excel('unemployment.xlsx', sheet_name= 'Sheet 1', skiprows = 10)
ds = pd.concat([une['TIME'],une.filter(like = '2019').mean(axis =1), une.filter(like = '2020').mean(axis =1)], axis =1)
ds['change'] = ds[1] - ds[0]
ds
```

### GDP 

```python
gdp = pd.read_excel('GDP.xlsx', sheet_name= 'Sheet 1', skiprows = 9)
gdp
ds = pd.concat([gdp['TIME'],gdp.filter(like = '2019').replace(':', 0).mean(axis =1), gdp.filter(like = '2020').replace(':', 0).mean(axis =1)], axis =1)
ds['change'] = ds[1] - ds[0]
ds
```
