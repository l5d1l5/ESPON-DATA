```python
import pandas as pd 
import os 
```

```python
files =  [file for file in os.listdir() if file.endswith('.xlsx')]

df = []

for file in files:
    read = pd.read_excel(file)
    df.append(read)
```

```python
ds = (pd.concat([pd.concat(df, axis =1).filter(like = '2019').mean(axis = 1), pd.concat(df, axis =1).filter(like = '2020').mean(axis = 1), pd.concat(df, axis =1).filter(like = '2021').mean(axis = 1)], axis =1) * 100)
ds = pd.concat([df[0].iloc[:,0], ds], axis =1)
ds.columns = ['name', '2019', '2020', '2021']
ds
```

```python
ds.to_excel('SE-totale.xlsx', index = False)
```
