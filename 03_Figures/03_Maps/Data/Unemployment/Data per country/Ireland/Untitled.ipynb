```python
import pandas as pd 
```

```python
s = pd.read_excel('Ireland.xlsx', sheet_name = 'Unpivoted').drop(['UNIT', 'STATISTIC'], axis =1)
s['Quarter'] = s['Quarter'].apply(lambda x: x[:4])
s.groupby(['Quarter', 'Region']).mean().unstack().T.reset_index().drop('level_0', axis =1).to_csv('Ireland-done.csv')
```
